```python
df=pd.read_csv('/lulus.csv')
df
```

```python
df.rename(columns={'Lulus Cepat':'Kelulusan'},inplace=True)
df
```

```python
df['Kelulusan'] = df['Kelulusan'].replace([1], 'Lulus Cepat')
df['Kelulusan'] = df['Kelulusan'].replace([0], 'Lulus Biasa')
df
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for kl, d in df.groupby('Kelulusan'):
    ax.scatter(d['IPK'], d['Prestasi'], label=kl)

plt.legend(loc='upper left')
plt.title('Sebaran Data IPK, Prestasi, dan Kelulusan')
plt.xlabel('IPK')
plt.ylabel('Prestasi')
plt.grid(True)
plt.show()
```

```python
import numpy as np

x_train = np.array(df[['IPK', 'Prestasi']])
y_train = np.array(df['Kelulusan'])

print(f'X_train:\n{x_train}\n')
print(f'y_train: {y_train}')
```

```python
from sklearn.preprocessing import LabelBinarizer

from sklearn.preprocessing import LabelBinarizer
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
print(f'y_train:\n{y_train}')
```

```python
y_train = y_train.flatten()
print(f'y_train: {y_train}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

K = 3

model = KNeighborsClassifier(n_neighbors=K)

model.fit(x_train, y_train)
```

```python
x_test = np.array([[3.12, 2], [3.75, 1], [2.88, 0], [2.79, 1]])
y_test = lb.transform(np.array(['Lulus Cepat', 'Lulus Cepat', 'Lulus Biasa', 'Lulus Cepat'])).flatten()

print(f'X_test:\n{x_test}\n')
print(f'y_test:\n{y_test}')
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)

print(f'Accuracy: {acc}')
```

```python
from sklearn.metrics import precision_score

prec = precision_score(y_test, y_pred)

print(f'Precision: {prec}')
```

```python
from sklearn.metrics import recall_score

rec = recall_score(y_test, y_pred)

print(f'Recall: {rec}')
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print(f'F1 Score: {f1}')
```

```python
from sklearn.metrics import classification_report

cls_report = classification_report(y_test, y_pred)

print(f'Classification Report:\n{cls_report}')
```

```python
from sklearn.metrics import matthews_corrcoef

mcc = matthews_corrcoef(y_test, y_pred)

print(f'MCC: {mcc}')
```
